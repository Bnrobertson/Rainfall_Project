```python
# Import dependencies
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
# Read in CSV
df = pd.read_csv("Resources/Farmer's_MarketsCSV.csv")

# Display data
df.head()
```

```python
df.info()
```

```python
# Dropping columns we don't need
df = df.drop(["OBJECTID","PLANNING_JURISDICTION", "ZONING_DISTRICT", 
              "F__OF_VENDORS", "GOODNESS_GROWS", "SPECIAL_SERVICES", "PRODUCTS"], axis=1)

# Display data
df.head()
```

```python
# Renaming columns
df = df.rename(columns={
    "MARKET_NAME": "Market Name",
    "ADDRESS": "Address",
    "CITY": "City",
    "PHONE": "Phone",
    "WEBSITE": "Website",
    "SEASONS_OF_OPERATION": "Seasons Open",
    "DAYS_OF_OPERATION": "Days Open",
    "TIME_OF_OPERATION": "Hours of Operation",
    "ACCEPT_EBT": "Accepts EBT"
}
                 )
# Display data
df.head()
```

```python
df.fillna('N/A', inplace=True)
```

```python
df.info()
```

```python
# Adding the Market Name to the index
df.set_index("Market Name", inplace=True)

# Display data
df.head()
```

```python
# Convert to CSV
df.to_csv("Farmers_Market_Clean.csv")
```
