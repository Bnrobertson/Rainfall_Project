```python
# Import dependencies
import pandas as pd
from meteostat import Point, Daily
from datetime import datetime
import numpy as np
```

```python
# Pull dates from Meteo website
start = datetime(2000, 1, 1)
end = datetime(2024, 5, 17)

# Raleigh, NC
raleigh = Point(35.7796, -78.6382)
# Fetch data
data = Daily(raleigh, start, end)
data = data.fetch()

# Printing data
print(data)
```

```python
# Convert into a DataFrame
df = pd.DataFrame(data)
```

```python
# Display top rows of DataFrame
df.head()
```

```python
# Display bottom rows of DataFrame
df.tail()
```

```python
# Summary statistics for numerical columns in our DataFrame
df.info()
```

```python
# Checking for Null values
print(df.isna().any().any())
```

```python
# Changing prcp into an interger
df["prcp"] = df["prcp"].astype(int)

# Display data
df.head()
```

```python
# Reseting the index and changing time to datetime format
df.reset_index(inplace=True)
df["time"] = pd.to_datetime(df["time"]) 
df.info()
```

```python
# Dropping columns we don't need
df = df.drop(["wpgt","tsun", "wdir", "snow"], axis=1)

# Display data
df.head()
```

```python
# Renaming columns
df = df.rename(columns={
    "time": "Date",
    "tavg": "TempAvgF",
    "tmin": "TempMinF",
    "tmax": "TempMaxF",
    "prcp": "RainTotal",
    "wspd": "WindSpeed",
    "pres": "Pressure",
    "rain": "Raining"
}
                 )
# Display data
df.head()
```

```python
# Summary statistics for numerical columns in our DataFrame
df.info()
```

```python
# Display data
df.head()
```

```python
# Display bottom rows of data
df.tail()
```

```python
# Converting Temp columns to Fahrenheit 
df["TempAvgF"] = df["TempAvgF"] * 9/5 + 32
df["TempMinF"] = df["TempMinF"] * 9/5 + 32
df["TempMaxF"] = df["TempMaxF"] * 9/5 + 32

# Converting RainTotal to inches
df["RainTotal"] = df["RainTotal"] * 0.0393701

# Converting Pressure to inHg
df["Pressure"] = df["Pressure"] * 0.02953

# Convert Windspeed to mph
df["WindSpeed"] = df["WindSpeed"] * 0.621371


# Display data
df.head()
```

```python
# # Converting RainTotal to inches
# df["RainTotal"] = df["RainTotal"] * 0.0393701

# # Display data
# df.head()
```

```python
# Dropping null values
df_cleaned = df.dropna()

# Display data
df_cleaned
```

```python
# Changing Raintotal into bool type
df_cleaned["Raining"] = (df_cleaned["RainTotal"] > 0).astype(bool)


target = df_cleaned["Raining"]

# Display data
df_cleaned.head()
```

```python
# Adding the Date to the index
df_cleaned.set_index("Date", inplace=True)

# Display data
df_cleaned.head()
```

```python
# Summary statistics for numerical columns in our DataFrame
df_cleaned.info()
```

```python
# Convert to CSV
df_cleaned.to_csv("Rainfall_OG.csv")
```
