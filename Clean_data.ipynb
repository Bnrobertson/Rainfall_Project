```python
import pandas as pd
from meteostat import Point, Daily
from datetime import datetime
import numpy as np
```

```python
start = datetime(2000, 1, 1)
end = datetime(2024, 5, 17)

# Raleigh, NC
raleigh = Point(35.7796, -78.6382)

data = Daily(raleigh, start, end)
data = data.fetch()


print(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
print(df.isna().any().any())
```

```python
df["prcp"] = df["prcp"].astype(int)
df.head()
```

```python
# df["time"] = pd.to_datetime(df.index["time"]).reset_index(drop=False)
# df.index = pd.to_datetime(df.index)
# df.info()
# df = df.drop(df.index[[8904, 8905, 8906]]).reset_index(drop=True)
# df.tail()
df.reset_index(inplace=True)
df["time"] = pd.to_datetime(df["time"]) 
df.info()
```

```python
df = df.drop(["wpgt","tsun", "wdir", "snow"], axis=1)
df.head()
```

```python
df = df.rename(columns={
    "time": "Date",
    "tavg": "TempAvgºF",
    "tmin": "TempMinºF",
    "tmax": "TempMaxºF",
    "prcp": "RainTotal",
    "wspd": "WindSpeed",
    "pres": "Pressure",
    "rain": "Raining"
}
                 )

df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df["TempAvgºF"] = df["TempAvgºF"] * 9/5 + 32
df["TempMinºF"] = df["TempMinºF"] * 9/5 + 32
df["TempMaxºF"] = df["TempMaxºF"] * 9/5 + 32
df.head()
```

```python
df["RainTotal"] = df["RainTotal"] * 0.0393701
df.head()
```

```python
df["Pressure"] = df["Pressure"] * 0.02953
df.head()
```

```python
df["WindSpeed"] = df["WindSpeed"] * 0.621371
df.head()
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
df_cleaned["Raining"] = (df_cleaned["RainTotal"] > 0).astype(bool)

features = df_cleaned[["RainTotal", "WindSpeed", "Pressure"]]
target = df_cleaned["Raining"]
df_cleaned.head()
```

```python
df_cleaned.set_index("Date", inplace=True)

df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
df.to_csv("Rainfall_OG.csv")
```
