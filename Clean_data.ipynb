```python
df = df.drop(df.index[[8904, 8905, 8906]]).reset_index(drop=True)
df.tail()
```

```python
df.head()
```

```python
df["tavg"] = df["tavg"] * 9/5 + 32
df["tmin"] = df["tmin"] * 9/5 + 32
df["tmax"] = df["tmax"] * 9/5 + 32
df.head()
```

```python
df["prcp"] = df["prcp"] * 0.0393701
df.head()
```

```python
df["pres"] = df["pres"] * 0.02953
df.head()
```

```python
df = df.drop(["wpgt","tsun", "wdir", "snow"], axis=1)
df.head()
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
df_cleaned["rain"] = df_cleaned["rain"].astype(int)
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
df_cleaned = df_cleaned.rename(columns={
    "time": "Date",
    "tavg": "TempAvgÂºF",
    "tmin": "TempMinÂºF",
    "tmax": "TempMaxÂºF",
    "prcp": "RainTotal",
    "wspd": "WindSpeed",
    "pres": "Pressure",
    "rain": "Raining",
    "hummax": "HumidityMax",
    "humavg": "HumidityAvg",
    "hummin": "HumidityMin"
}
                 )

df_cleaned.head()
```

```python
df_cleaned["WindSpeed"] = df_cleaned["WindSpeed"] * 0.621371
df_cleaned.head()
```

```python
df_cleaned["Raining"] = df_cleaned["Raining"].astype(bool)
df_cleaned.head()
```
